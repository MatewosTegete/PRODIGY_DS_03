Data Preprocessing

```python
# Let's first check if any data are unavailable
bank_info.isna().sum()
```

```python
bank_info['housing'] = bank_info['housing'].replace('admin.', 'unknown')
bank_info['housing'] = bank_info['housing'].replace('blue-collar', 'unknown')
bank_info['housing'] = bank_info['housing'].replace('self-employed', 'unknown')
bank_info['housing'] = bank_info['housing'].replace('management', 'unknown')
A = bank_info['age']
b = bank_info['housing']

# for i in range(len(b)):
#     if type(b[i]) == str:
#       b = b.replace('unknown', 'no')

modified_data = []
def replace_unwanted_values(data):

  for val in data:
    val = val.lower()
    if val not in ("yes", "no"):
      modified_data.append('no')
    else:
      modified_data.append(val)
  return modified_data
b = replace_unwanted_values(b)
df = pd.DataFrame(b)
le = LabelEncoder()
b_encoded = le.fit_transform(df)

```

```python
# For Age and Housing data to guess Housing existence based on age
# Split data into training and testing sets
A_train, A_test, b_train, b_test = train_test_split(A, b_encoded, test_size=0.2)
A_train = np.array(A_train).reshape(-1, 1)
A_test = np.array(A_test).reshape(-1, 1)
# Create and fit the decision tree model
model = DecisionTreeClassifier()

model.fit(A_train, b_train)

predictions = model.predict(A_test)

# Evaluate the model accuracy
score = accuracy_score(b_test, predictions)

print(f"Model accuracy: {score:.2f}")
```

So, We can say the co-relation is slightly better than random
